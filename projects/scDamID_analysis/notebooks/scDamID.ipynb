<h2> Create individual bedgraphs from scDamID data

```python
import numpy as np
import pandas as pd
```

```python
inputfile=open("DataS2_Clone.5-5.1N.OE.txt", 'r')
```

```python
df=pd.read_csv("DataS2_Clone.5-5.1N.OE.txt", delimiter='\t')
```

```python
df.head()
```

```python
#Create Individual bedgraphs
for i,c in enumerate(df.columns[3:]):
    cols_to_keep=['seqnames','start','end',c]
    df.to_csv(c+'.bedgraph', sep='\t', header=False, columns=cols_to_keep, index=False)
```
